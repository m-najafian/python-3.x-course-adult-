```python
import matplotlib.pyplot as plt
```

```python
import csv

x = []
y = []

with open("files/ex06.txt", "r") as csvFile:
    plots = csv.reader(csvFile, delimiter=",")
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x, y, label="Data From File")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Main Plot")
plt.legend()
plt.show()
```

```python
import csv

x = []
y = []

with open("files/ex06-1.txt", "r") as csvFile:
    plots = csv.reader(csvFile, delimiter=",")
    all_data = []
    for row in plots:
        all_data.append(row)
    x = all_data[0]
    y = all_data[1]
for item in range(len(x)):
    x[item]=int(x[item])

plt.pie(x,labels=y)
plt.show()
```

```python
import numpy as np

x, y = np.loadtxt("files/ex06.txt", delimiter=",", unpack=True)

plt.plot(x, y, label="Data From File")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Main Plot")
plt.legend()
plt.show()
```
