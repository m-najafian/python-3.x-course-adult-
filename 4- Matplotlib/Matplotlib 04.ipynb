```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [2,3,4,5,1,6,7,4]

plt.scatter(x,y, label="MyScats")

plt.xlabel('X')
plt.ylabel('Y')
plt.title('My Main Graph')
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [2,3,4,5,1,6,7,4]

plt.scatter(x,y, label="MyScats", color="k", marker="*", s=100)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('My Main Graph')
plt.legend()
plt.show()
```
