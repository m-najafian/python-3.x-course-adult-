```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]

working = [2,8,3,5,8]
eating  = [1,6,5,8,6]
sleeping= [8,5,8,3,9]
playing = [5,5,2,7,8]

plt.stackplot(days, working,eating,sleeping,playing, colors=["c","m","r","k"])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('My Main Graph')
plt.legend()
plt.show()
```
