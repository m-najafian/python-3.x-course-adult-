```python
import matplotlib.pyplot as plt
```

```python
slices = [21,6,3,7,13]
activities = ["Python","PHP","JAVA","C++","Swift"]

plt.pie(slices, labels=activities)

plt.title("Main Plot")
plt.show()
```

```python
slices = [21,6,3,7,13]
activities = ["Python","PHP","JAVA","C++","Swift"]
ac_colors = ["c","r","yellow","navy","orange"]

plt.pie(slices, labels=activities, colors=ac_colors)

plt.title("Main Plot")
plt.show()
```

```python
slices = [21,6,3,7,13]
activities = ["Python","PHP","JAVA","C++","Swift"]
ac_colors = ["cornflowerblue","coral","forestgreen","crimson","blueviolet"]

plt.pie(slices, labels=activities, colors=ac_colors, startangle=90)

plt.title("Main Plot")
plt.show()
```

```python
slices = [21,6,3,7,13]
activities = ["Python","PHP","JAVA","C++","Swift"]
ac_colors = ["cornflowerblue","coral","forestgreen","crimson","blueviolet"]

plt.pie(slices, labels=activities, colors=ac_colors, startangle=90, shadow=True)

plt.title("Main Plot")
plt.show()
```

```python
slices = [21,6,3,7,13]
activities = ["Python","PHP","JAVA","C++","Swift"]
ac_colors = ["cornflowerblue","coral","forestgreen","crimson","blueviolet"]

plt.pie(slices, labels=activities, colors=ac_colors, startangle=90, explode=(0,0,0.1,0,0))

plt.title("Main Plot")
plt.show()
```

```python
slices = [21,6,3,7,13]
activities = ["Python","PHP","JAVA","C++","Swift"]
ac_colors = ["cornflowerblue","coral","forestgreen","crimson","blueviolet"]

plt.pie(slices, labels=activities, colors=ac_colors, startangle=90, explode=(0,0,0.1,0,0), autopct="%1.1f%%")

plt.title("Main Plot")
plt.show()
```
