```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [x + 2000 for x in range(20)]
y = [2,6,4,1,5,6,7,3,4,5,6,2,8,9,9,8,8,9,9,6]

plt.figure(figsize=(20,5))
plt.plot(x,y,label="Price")
plt.xticks(rotation=45)

plt.show()
```
